```python
df = spark.read.csv(
    "Files/Bronze/Hospital_General_Information.csv",
    header=True,
    inferSchema=True
)

# Replace spaces with underscores
for col in df.columns:
    df = df.withColumnRenamed(col, col.replace(" ", "_").replace("/", "_"))

df.write.mode("overwrite").saveAsTable("bronze_hospital_information")
```
